```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import datasets, preprocessing
from sklearn.cluster import KMeans
```

```python
data = pd.read_csv(r"Test_Data.csv")
data.head()
```

```python
data_frame = pd.DataFrame(data)
```

```python
print(data_frame.head())
```

```python
Main = data_frame[["STG","PEG"]]
s = preprocessing.StandardScaler()
s.fit(Main)
Main_norm = s.transform(Main)
```

```python
cls = KMeans(n_clusters = 4)
result = cls.fit(Main_norm)
plt.scatter(Main_norm[:,0],Main_norm[:,1], c=result.labels_)
plt.scatter(result.cluster_centers_[:,0],result.cluster_centers_[:,1],s=250, marker='*',c='red')
#plt.set_xlabel("STGの値を標準化した値",size = 14)
#plt.set_ylabel("PEGの値を標準化した値",size = 14)
plt.show()
```

```python
Main_t = data_frame[["STR","LPR"]]
s = preprocessing.StandardScaler()
s.fit(Main_t)
Main_t_norm = s.transform(Main_t)
```

```python
cls = KMeans(n_clusters = 4)
result = cls.fit(Main_t_norm)
plt.scatter(Main_t_norm[:,0],Main_t_norm[:,1], c=result.labels_)
plt.scatter(result.cluster_centers_[:,0],result.cluster_centers_[:,1],s=250, marker='*',c='red')
plt.show()
```
